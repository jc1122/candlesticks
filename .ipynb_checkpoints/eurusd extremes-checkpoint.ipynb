## Import data

Import hourly data from CSV and prepare time series.

```python
import pandas as pd
# import data
hourly_df = pd.read_csv("EURUSD60.csv",
                        names=["Date","Time","Open","High","Low","Close","Volume"],
                       )
hourly_df['DateTime'] = pd.to_datetime(hourly_df['Date']
                                       + ' ' + hourly_df['Time']) #join separate columns date + time and convert to datetime
hourly_df.drop(columns=['Date', 'Time'], inplace=True) # drop redundant columns
hourly_df.head()
```

## Investigate the time of occurence of daily extremes

Is there an icreased probability of daily extremes to occur on given hour?

### Prepare data

Separate each day as an index.

```python
# recreate date and time and make multiindex
hourly_df['Date'] = hourly_df['DateTime'].apply(lambda x: x.date()) # extract date fom Timestamp
hourly_df['Time'] = hourly_df['DateTime'].apply(lambda x: x.time()) # extract time from Timestamp
hourly_df.set_index(['Date','Time'],inplace=True) # create multiindex
hourly_df.drop(columns=['DateTime'], inplace=True) # drop redundant column
hourly_df.head(5)
#hourly_df = hourly_df[:pd.to_datetime('2014')]
```

### Find time of occurence of daily extreme prices

```python
#occurence of high for day:
extremes = pd.DataFrame(hourly_df.groupby(level='Date')['High'].apply(lambda x: x.idxmax()[1]))
extremes.columns = ['HighTime']
#occurence of low for day:
extremes['LowTime'] = pd.DataFrame(data=hourly_df.groupby(level='Date')['Low'].apply(lambda x: x.idxmin()[1]))
```

```python
extremes.head()
```

### Count occurences of daily extremes on given hour

```python
def counts(extremes_df):
    extremes_counts = pd.DataFrame(index=extremes_df.groupby('LowTime')['LowTime'].count().index)
    extremes_counts.index.names = ['Time']

    extremes_counts['LowCount'] = extremes_df.groupby('LowTime')['LowTime'].count()
    extremes_counts['HighCount'] = extremes_df.groupby('HighTime')['HighTime'].count()
    return extremes_counts
```

```python
# create a DataFrame for couts
extremes_counts = counts(extremes)

extremes_counts.head(5)
```

```python
# Are the highs and lows likely to be from the same distribution?
from scipy import stats
stats.ks_2samp(extremes_counts['HighCount'], extremes_counts['LowCount'])
```

With p value of 0.98, we cannot reject the hypothesis, that both distributions are the same. Then it is appropriate to aggregate the values.

Check if both series are correlated:

```python
print ("correlation: " + str(extremes_counts['LowCount'].corr(extremes_counts['HighCount'])))
```

### Plot a histogram of counts of daily extremes

See plot, to know at which hour the daily extremes are most likely to occur. At top of the bars is % probability of extreme to occur at given hour. The values in the middle of colored bars show the count of extremes which occured at given hour. 

```python
def plot_counts(extremes_counts):
    # prepare bar plot
    ax = extremes_counts.plot.bar(stacked=True, figsize=(20,4))
    ax.set_title('Counts and probability of ocurence of daily extreme on given hour',fontsize=24)
    # annotate lower bars with counts at 1/2 height of the bar
    for p in ax.patches[:24]:
        ax.annotate(str(p.get_height()), xy=(p.get_x(), p.get_height()/2))

    # annotate upper bars with counts at 1/2 height of the bar    
    i=0
    for p in ax.patches[24:]:
        ax.annotate(str(p.get_height()), xy=(p.get_x(), p.get_height()/2 + ax.patches[i].get_height()))
        i +=1

    # annotate total height with probability of occurence of extreme at top of sum of the bars    
    sum_of_extremes = extremes_counts.sum().sum()
    for i in range(0,24):
        height = ax.patches[i].get_height() + ax.patches[i+24].get_height()
        ax.annotate("{0:.2f}%".format(height/sum_of_extremes*100),xy=(ax.patches[i].get_x(),height))
```

```python
%matplotlib inline

plot_counts(extremes_counts)
```

Under assumption, that the time of occurences of daily extremes are independent, it is most likely, that one extreme will occur at 00:00 - 01:00 and the second on 16:00 - 17:00.

### Plot a heatmap of pairs of occurences of daily extremes time

Prepare data for heatmap. Use extreme 2 time as columns and extreme 1 time as rows.

```python
def aggregate_extremes(extremes):
    #treat high time and low time equally as extreme time and remove duplicates:
    extremes_aggregated = extremes.copy(deep=True)
    extremes_aggregated.loc[extremes_aggregated['HighTime'] > extremes_aggregated['LowTime'],['HighTime', 'LowTime']] = \
    extremes_aggregated.loc[extremes_aggregated['HighTime'] > extremes_aggregated['LowTime'],['LowTime', 'HighTime']].values

    #unstack the index to have a Dataframe with time as index and column
    extremes_occurence = extremes_aggregated.groupby(['LowTime','HighTime'])['LowTime'].count().unstack(level=0)
    extremes_occurence.index.name = 'Extreme1 Time'
    extremes_occurence.columns.name = 'Extreme2 Time'

    #convert to probability by dividing each cell by sum of all cells
    extremes_occurence = extremes_occurence.div(extremes_occurence.sum().sum())
    return extremes_occurence
```

```python
extremes_occurence = aggregate_extremes(extremes)
extremes_occurence.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def extremes_heatmap(extremes_occurence):
    %matplotlib inline
    fig, ax = plt.subplots(figsize=(20,10))
    ax.set_title('Probability of occurrence of extremes at given time')
    sns.heatmap(extremes_occurence, annot=True, ax=ax, fmt='.2%')
```

```python
extremes_heatmap(extremes_occurence)
```

### Plot histogram and heatmap of occurences of extremes during winter and summer time

## What is the probability of new extreme during the day, given actual time and time of actual extremes?

```python
# the arguments take ints not column names!!! this will give errors in the future!!!!! refactor
def probability_of_new_extreme(extremes_occurence, extreme1_time, extreme2_time, actual_time):
    prob = 0
    prob += extremes_occurence.loc[extremes_occurence.index[extreme1_time]:extremes_occurence.index[extreme1_time],
                       extremes_occurence.columns[actual_time-1]:extremes_occurence.columns[22]].sum().sum()
    prob += extremes_occurence.loc[extremes_occurence.index[extreme2_time]:extremes_occurence.index[extreme2_time],
                       extremes_occurence.columns[actual_time-1]:extremes_occurence.columns[22]].sum().sum()
    prob += extremes_occurence.loc[extremes_occurence.index[actual_time-1]:extremes_occurence.index[22],
                       extremes_occurence.columns[actual_time-1]:extremes_occurence.columns[22]].sum().sum()
    return prob
```
